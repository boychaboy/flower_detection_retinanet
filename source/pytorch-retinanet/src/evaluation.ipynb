# 실험 결과 시각화
## 1. 파일 읽어들이기

```python
def read_file(filename):
    with open(filename, 'r') as file:
        lines = file.readlines()
        epochs = []
        flower_acc = []
        for e in range(0,len(lines),2):
            epoch = lines[e].strip().split()[1:]
            flower_acc = lines[e+1].strip().split()
            epoch.append(flower_acc)
            epochs.append(epoch)
    return epochs

epochs_18 = read_file('21class_18_result.txt')
epochs_50 = read_file('21class_50_result.txt')
epochs_101 = read_file('21class_101_result.txt')
epochs_152 = read_file('21class_152_result.txt')
epochs10_152 = read_file('10class_152_result.txt')
```

```python
epochs_50
```

```python
import numpy as np

loss_val_50 = []

for epoch in epochs_50:
    new = [float(epoch[0]),float(epoch[1])]
    loss_val_50.append(new)

loss_val_101 = []

for epoch in epochs_101:
    new = [float(epoch[0]),float(epoch[1])]
    loss_val_101.append(new)

loss_50 = np.array(loss_val_50)[:,0]
loss_50 = loss_50.tolist()
val_50 = np.array(loss_val_50)[:,1].tolist()
loss_101 = np.array(loss_val_101)[:,0].tolist()
val_101 = np.array(loss_val_101)[:,1].tolist()

loss_val_18 = []

for epoch in epochs_18:
    new = [float(epoch[0]),float(epoch[1])]
    loss_val_18.append(new)

loss_val_152 = []

for epoch in epochs_152:
    new = [float(epoch[0]),float(epoch[1])]
    loss_val_152.append(new)

loss_18 = np.array(loss_val_18)[:,0].tolist()
val_18 = np.array(loss_val_18)[:,1].tolist()
loss_152 = np.array(loss_val_152)[:,0].tolist()
val_152 = np.array(loss_val_152)[:,1].tolist()

loss_val10_152 = []
for epoch in epochs10_152:
    new = [float(epoch[0]),float(epoch[1])]
    loss_val10_152.append(new)

loss10_152 = np.array(loss_val10_152)[:,0].tolist()
val10_152 = np.array(loss_val10_152)[:,1].tolist()
```

## 2. Loss시각화

```python
import matplotlib.pyplot as plt


def visualize(list1, list2, list3, list4, text):
#     plt.style.use(['seaborn'])
    plt.figure(figsize=(20,15))

    step = len(list1)
    plt.plot(np.arange(step), list1, marker='.',label = 'ResNet18')
    plt.plot(np.arange(step), list2, marker='.',label = 'ResNet50')
    plt.plot(np.arange(step), list3, marker='.',label = 'ResNet101')
    plt.plot(np.arange(step), list4, marker='.',label = 'ResNet152')
    
    plt.xlabel('Epoch', fontsize=20)
    plt.ylabel(text, fontsize=20)
    plt.title(text, fontsize=30)
    plt.xticks(range(0, step, 1), fontsize=20)
    plt.yticks(fontsize=20)
    plt.legend(['ResNet18','ResNet50','Resnet101','ResNet152'], fontsize=20)
    plt.show()

visualize(loss_18, loss_50, loss_101, loss_152, 'Loss')
visualize(val_18, val_50, val_101, val_152, 'Accuracy')
```

```python
def visualize2(list1, list2, text):
#     plt.style.use(['seaborn'])
    plt.figure(figsize=(20,15))

    step = len(list1)
    plt.plot(np.arange(step), list1, marker='.',label = '10 class')
    plt.plot(np.arange(step), list2, marker='.',label = '21 class')
    
    plt.xlabel('Epoch', fontsize=20)
    plt.ylabel(text, fontsize=20)
    plt.title(text, fontsize=30)
    plt.xticks(range(0, step, 1), fontsize=20)
    plt.yticks(fontsize=20)
    plt.legend(['10 class','21 class'], fontsize=20)
    plt.show()

visualize2(loss10_152, loss_152, 'Loss')
visualize2(val10_152, val_152, 'Accuracy')
```
